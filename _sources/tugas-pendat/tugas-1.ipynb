# Tugas 1 Dissimilarity

```python
import numpy as np
import pandas as pd
import seaborn as sb
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/achmadbaharuddinakbar/datamining/main/StudentsPerformance.csv")
print(data)
```

```python
data
```

```python
def dt1(i,j):
  P = 9
  Q = 0
  for col in ["gender", "lunch", "test preparation course", "math score", "reading score", "writing score"]:
    if data[col][i] == data[col][j]:
      Q += 1
  return (P-Q)/P
```

```python
print(dt1(1,2))
print(dt1(1,3))
print(dt1(1,4))
```

```python
def dt2(i,j,h):
  result = 0
  for col in ["math score", "reading score","writing score"]:
    result = math.pow(abs(data[col][i] - data[col][j]), h)
  return result**(1/h)
```

```python
print(dt2(1,2,1))
print(dt2(1,3,1))
print(dt2(1,4,1))
```
